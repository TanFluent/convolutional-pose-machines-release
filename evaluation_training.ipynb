--针对 caffe_cpm 官方版本的输出log格式进行解析

```python
# libs
import os
import pdb

import numpy as np

# dirs
training_log_file_path = './caffe_log_7108.log'
```

```python
# Log file Parsering
lgf = open(training_log_file_path,'r')

lines = lgf.readlines()

# information
batch_num = 0
batch_loss = 0
avg_loss = 0
lr = 0

stage_batch_loss_dict = {}

batch_num_list = []
batch_loss_list = []
lr_list = []


# parse info
cnt = 0 # flag for diff stage
for idx,line in enumerate(lines):
    
    data = line.strip().split(' ')
    
    if ' Iteration ' in line:
        
        
        if 'loss =' in line:
            batch_num = data[6]
            batch_loss = float(data[-1])
            cnt = 0 # batch info start
            loss_dict = {}
            
        if 'lr =' in line:
            lr = float(data[-1])
            batch_num_list.append(batch_num)
            batch_loss_list.append(batch_loss)
            lr_list.append(lr)
            
            for k,v in loss_dict.iteritems():
                if stage_batch_loss_dict.has_key(k):
                    stage_batch_loss_dict[k].append(v)
                else:
                    stage_batch_loss_dict[k] = [v]
            
    if ('loss_stage' in line) and ('Train net output #' in line):
        loss_dict[data[-8]] = float(data[-6])

        
#pdb.set_trace()
print('Log file parsering done!')
```

```python
# Plot
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.plotly as py
import plotly.graph_objs as go

# --show image on jupyter notebook
init_notebook_mode(connected=True)

# --Create traces
trace0 = go.Scatter(
    x = np.array(batch_num_list),
    y = batch_loss_list,
    mode = 'lines',
    name = 'lines'
)

py_data = [trace0]

iplot(py_data, filename='line-mode',show_link=False)
```
